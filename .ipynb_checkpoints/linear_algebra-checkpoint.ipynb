```python
#Eigen decomposition
from numpy import linalg
```

```python
#Define an array
A = np.arange(9) - 3
A
```

```python
B = A.reshape((3,3))
B
```

```python
#Eucedian (L2) from
print(np.linalg.norm(A))
print(np.linalg.norm(B))
```

```python
#forbeniusm norm
print(np.linalg.norm(B, 'fro'))
```

```python
#Max norm (P = infinity)
print(np.linalg.norm(A, np.inf))
print(np.linalg.norm(B, np.inf))
```

```python
#vector normalization - normalization to produce a unit vector
norm = np.linalg.norm(A)
A_unit = A / norm
print(A_unit)
```

```python
#the magnitude of a unit vector is equal to 1
np.linalg.norm(A_unit)
```

```python
#Eigendecompistion

#find the eignevalues and eigenvectors for a simple square matrix
A = np.diag(np.arange(1,4))
A
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
```

```python
eigenvalues
```

```python
eigenvectors
```

```python
#verify eigendecomposition
matrix = np.matmul(np.diag(eigenvalues), np.linalg.inv(eigenvectors))
output = np.matmul(eigenvectors, matrix).astype(int)
print(output)
```

```python
#import necessary matplotlib libraries
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#plot the eigenvectors
origin = [0,0,0]

fig = plt.figure(figsize=(18,10))
fig.suptitle('Effects of eigenvalues and eigenvectors')
axl = fig.add_subplot(121, projection = '3d')

axl.quiver(origin, origin, origin, eigenvectors[0, :], eigenvectors[1, :], eigenvectors[2, :], color = 'k')
axl.set_xlim([-3, 3])
axl.set_ylim([-3, 3])
axl.set_zlim([-3, 3])
axl.set_xlabel('X axis')
axl.set_ylabel('Y axis')
axl.set_zlabel('Z axis')
axl.view_init(15, 30)
axl.set_title('Befor Multiplication')

#show the plot
plt.show()
```

```python
#plot the eigenvectors
origin = [0,0,0]

fig = plt.figure(figsize=(18,10))
fig.suptitle('Effects of eigenvalues and eigenvectors')
axl = fig.add_subplot(121, projection = '3d')

axl.quiver(origin, origin, origin, eigenvectors[0, :], eigenvectors[1, :], eigenvectors[2, :], color = 'k')
axl.set_xlim([-3, 3])
axl.set_ylim([-3, 3])
axl.set_zlim([-3, 3])
axl.set_xlabel('X axis')
axl.set_ylabel('Y axis')
axl.set_zlabel('Z axis')
axl.view_init(15, 30)
axl.set_title('Befor Multiplication')

#Multiply original matrix by eigenvectors
new_eig = np.matmul(A, eigenvectors)
axl2 = fig.add_subplot(121, projection = '3d')

axl2.quiver(origin, origin, origin, new_eig[0, :], new_eig[1, :], new_eig[2, :], color = 'k')

#add eignevalues to the plot
axl2.plot((eigenvalues[0]*eigenvectors[0]), (eigenvalues[1]*eigenvectors[1]), (eigenvalues[2]*eigenvectors[2]), 'rx') 
axl2.set_xlim([-3, 3])
axl2.set_ylim([-3, 3])
axl2.set_zlim([-3, 3])
axl2.set_xlabel('X axis')
axl2.set_ylabel('Y axis')
axl2.set_zlabel('Z axis')
axl2.view_init(15, 30)
axl.set_title('After Multiplication')

#show the plot
plt.show()
```
