```python
import sys
import numpy as np
import matplotlib 
import matplotlib .pyplot as plt

print(sys.version)
print(np.version.version)
print(matplotlib.__version__)
```

```python
# generate 2d mesh grid

nx, ny = (100,100)

x = np.linspace(0, 10, nx)
y = np.linspace(0, 10, ny)

xv, yv = np.meshgrid(x, y)
```

```python
# define a function to plot

def f(x,y):
    return x * (y**2)

# calculate z value for each x, y point
z = f(xv, yv)
```

```python
# make a colorplot to display the data

plt.figure(figsize=(14,12))
plt.pcolor(xv, yv, z)
plt.title("2D color plot of f(x,y) = xy^2")
plt.colorbar()
plt.show()
```

```python
# generate a 2D meshgrif for the grdient
nx, ny = (10,10)
x = np.linspace(0, 10, nx)
y = np.linspace(0, 10, ny)
xg, yg = np.meshgrid(x, y)

#calculate the grdient of f(x,y)
Gy, Gx = np.gradient(f(xg, yg)) # rows are y 
```

```python
# make a colorplot to display the data

plt.figure(figsize=(14,12))
plt.pcolor(xv, yv, z)
plt.title("Gradient of f(x,y) = xy^2")
plt.colorbar()
plt.quiver(xg, yg, Gx, Gy, scale = 1000, color = 'w')
plt.show()
```

```python
# calculate the gradient of f(x, y) =xy^2 (to verify the numpy ccaluclated correctyly)

def ddx(x,y):
    return y ** 2

def ddy(x,y):
    return 2 * x * y

Gx = ddx(xg, yg)
Gy = ddy(xg, yg)
```

```python
# make a colorplot to display the data

plt.figure(figsize=(14,12))
plt.pcolor(xv, yv, z)
plt.title("Plot of [y^2 , 2xy]")
plt.colorbar()
plt.quiver(xg, yg, Gx, Gy, scale = 1000, color = 'w')
plt.show()
```
